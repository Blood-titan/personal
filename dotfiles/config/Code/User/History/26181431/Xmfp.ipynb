```python
import pandas as pd
```

```python
movies = pd.read_csv("data/tmdb_5000_movies.csv")
movies.describe()
```

```python
import ast 
movies['geners_name'] = movies['genres'].apply( 
    lambda x: [d['name'] for d in ast.literal_eval(x)] if x != "[]" else {})
```

##create a dataset to train and test
